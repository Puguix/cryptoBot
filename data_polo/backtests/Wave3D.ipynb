<h1> Define Wave Trend 3D </h1>

```python
class WaveTrend3D:
    def __init__(self,
                df,
                oscillator_lookback=20,
                quadratic_mean_length=50,
                speed_to_emphasize='slow',
                emphasis_width=2,
                f_length=0.75,
                f_smoothing=0.45,
                n_length=1.0,
                n_smoothing=1.0,
                s_length=1.75,
                s_smoothing=2.5,
                div_threshold=30,
                size_percent=40):
    
        # Signal Settings
        self.df = df
        self.oscillator_loopback = oscillator_lookback
        self.quadratic_mean = quadratic_mean_length
        self.speed_to_emphasized = speed_to_emphasize
        self.emphasis_width = emphasis_width
        
        # Oscillator Settings
        self.fast_osc = f_length
        self.fast_osc_smooth = f_smoothing
        self.normal_osc = n_length
        self.normal_osc_smooth = n_smoothing
        self.slow_osc = s_length
        self.slow_osc_smooth = s_smoothing
        
        # Divergence Detection
        self.div_threshold = div_threshold
        self.size_percent = size_percent
        
        
```
